```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_txt(infile):
    # Path and filename of the text file
    input_file = str(infile)

    # Lists to store the data from each column
    tEnergy = []
    tZenith = []
    rEnergy = []
    rZenith = []
    charge = []

    # Reading the text file
    with open(input_file, 'r') as file:
        for line in file:
            # Splitting the line into columns based on the tab delimiter
            data = line.strip().split('\t')

            # Adding data to each column list
            tEnergy.append(float(data[0]))
            tZenith.append(float(data[1]))
            rEnergy.append(float(data[2]))
            rZenith.append(float(data[3]))
            charge.append(float(data[4]))
            
    
    return tEnergy, tZenith, rEnergy, rZenith, charge
```

```python
true_energy_RPE, true_zenith_RPE, reco_energy_RPE, reco_zenith_RPE, charge_RPE = read_txt('../txt_files/RPE_distributions.txt')
```

```python
true_energy, true_zenith, reco_energy, reco_zenith, charge = read_txt('../txt_files/distributions.txt')
```

```python
plt.figure(figsize=(10, 9))
plt.hist(np.log10(true_energy_RPE), bins=50, histtype='step', label='RPE')
plt.hist(np.log10(true_energy), bins=50, histtype='step', label='Original')
plt.legend()
plt.yscale('log')
plt.xlabel('True Energy[GeV]')
plt.title('NuMu')
```

```python
plt.figure(figsize=(10, 9))
plt.hist(np.log10(reco_energy_RPE), bins=50, histtype='step', label='RPE')
plt.hist(np.log10(reco_energy), bins=50, histtype='step', label='Original')
plt.legend()
plt.yscale('log')
plt.xlabel('Graphnet Reco Energy[GeV]')
plt.title('NuMu')
```

```python
plt.figure(figsize=(10, 9))
plt.hist(np.cos(np.array(true_zenith_RPE)), bins=50, histtype='step',label='RPE')
plt.hist(np.cos(np.array(true_zenith)), bins=50, histtype='step', label='Original')
plt.legend()
plt.xlabel('True Cos(Zenith)')
plt.title('NuMu')
```

```python
plt.figure(figsize=(10, 9))
plt.hist(np.cos(np.array(reco_zenith_RPE)), bins=50, histtype='step', label='RPE')
plt.hist(np.cos(np.array(reco_zenith)), bins=50, histtype='step', label='Original')
plt.legend()
plt.xlabel('Graphnet Reco Cos(Zenith)')
plt.title('NuMu')
```

```python
plt.figure(figsize=(10, 9))
plt.hist(charge_RPE, bins=50, histtype='step', label='RPE')
plt.hist(charge, bins=50, histtype='step', label='Original')
plt.legend()
plt.yscale('log')
plt.xlabel('L4 Fiducial Charge')
plt.title('NuMu')
```
