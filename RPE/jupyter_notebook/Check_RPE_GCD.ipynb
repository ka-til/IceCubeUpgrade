```python
from icecube import dataclasses, dataio, icetray
import matplotlib.pyplot as plt
import numpy as np
```

```python
gcd_infile = dataio.I3File('/mnt/c/Users/akank/IceCube_Software/i3/icetray/scripts/DOM_Eff/I3Files/GeoCalibDetectorStatus_ICUpgrade_rpe_v00.v58.mixed.V0.i3.bz2')
```

```python
while gcd_infile.more():
    frame = gcd_infile.pop_frame()
    if "I3Geometry" in frame:
        geometry = frame["I3Geometry"]

    if "I3Calibration" in frame:
        calibration = frame["I3Calibration"]
        calibrationData = calibration.dom_cal
        my_cal = []
        for omkey in geometry.omgeo.keys():
            oKey = geometry.omgeo.get(omkey)
            if omkey in calibrationData.keys():
                if calibrationData[omkey].relative_dom_eff < 0.989:
                    print(omkey, oKey.omtype, calibrationData[omkey].relative_dom_eff)
                else:
                    my_cal.append(calibrationData[omkey].relative_dom_eff)
```

```python
my_cal=np.array(my_cal)
```

```python
print(np.mean(my_cal[my_cal<1.35]), np.std(my_cal[my_cal<1.35]))
```

```python
plt.figure(figsize=(10,9))
plt.hist(my_cal[my_cal<1.35], bins=40, histtype = 'step', color = 'Red', linewidth=5)
plt.grid()
plt.title('RPE from GCD file - v00')
```
